# 5

```python
car_sales['Price'] = car_sales['Price'].str.replace('[$,]', '', regex= True).astype('int64')

car_sales.head()
```

# 6

```python
car_sales['Total Sales']= car_sales['Price'].cumsum()
```

```python
car_sales
```

# 7

```python
num_rows= len(car_sales)

start_date = datetime.now().date()
date_range = pd.date_range(start=start_date, periods=num_rows)

car_sales['Sale Date'] = date_range
```

```python
car_sales
```

# 8

```python
plt.figure(figsize=(8, 4))
plt.plot(car_sales['Price'], car_sales['Total Sales'])

plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs. Total Sales')

plt.show()
```

# 9

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.title('Odometer vs Price Scatter Plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

# 10

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'] )
plt.xlabel('Make')
plt.ylabel('Mean Odometer (KM)')
plt.title('Mean Odometer (KM) by Make')
plt.show()
```

# 11

```python
plt.hist(car_sales['Make'])
plt.title('Histogram of car make')
plt.xlabel('Make')
plt.show()
```

# 12

```python
fig=plt.figure(figsize=(14,7))

ax1=fig.add_subplot(231) 
ax1.set_title('Price vs. Total Sales Line Plot')
ax1.plot(car_sales['Total Sales'], car_sales['Price'])

ax2=fig.add_subplot(232) 
ax2.set_title('Scatter Plot of Odometer vs Price')
ax2.plot(car_sales['Odometer (KM)'],car_sales['Price'])

ax3=fig.add_subplot(233) 
ax3.set_title('Mean Odometer (KM) by Make')
ax3.plot(car_sales['Make'],car_sales['Odometer (KM)'])

ax4=fig.add_subplot(234) 
ax4.set_title('Distribution of Car Makes')
ax4.plot(car_sales['Make'])

plt.show()
```

# 16

```python
#Heatmap
heat_map= car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
cols= ['Total Sales', 'Price', 'Odometer (KM)']
sns.heatmap(heat_map,xticklabels=cols, yticklabels=cols, annot=True)
plt.show()
```

#### The correlation between and the odometer is negative and weak, while the correlation between total sales and price is positive and weak. The correlation between price and the odometer has a strong negative correlation.
